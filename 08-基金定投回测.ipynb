```python
import xalpha as xa
import pandas as pd

#易方达沪深300ETF联接A 110020
hs300 = xa.fundinfo('110020')

hs300.info()
```

```python
hs300_index = xa.indexinfo('0000300')

hs300_index.price[hs300_index.price['date'] >= '2020-03-01' ]
```

```python
#一年内对沪深300做网格交易
status = xa.policy.grid(hs300,[2,2,2,2,2],[3,3,3,3,3],'2019-03-01','2020-03-01',10000)

status.status
```

```python
#开始交易
trade = xa.trade(hs300,status.status)

trade.xirrrate('2020-03-01')
```

```python
trade.v_totvalue()
```

```python
trade.v_tradevolume()
```

```python
#每隔2周定投一次
data_range = pd.date_range('2019-03-07','2020-03-07',freq='14D')

#以定投的方式投资沪深300
dingtou = xa.policy.scheduled(hs300,500,data_range)

#开始实盘回测
shipan = xa.trade(hs300,dingtou.status)

#实盘收益率
shipan.xirrrate()
```

```python
import pandas as pd

ym = '2019-10'  #统计年月
rate = 6.50899   #当月汇率

path_ch = 'C:/Users/小奔/Desktop/编程练习/Python/数据分析/learnPython/data/ch.csv'
path_jp = 'C:/Users/小奔/Desktop/编程练习/Python/数据分析/learnPython/data/jp.csv'

#导入数据
df_ch = pd.read_csv(path_ch,
                    names=['id','input/output','type','money','datetime','comment','picture']
                    )

df_jp = pd.read_csv(path_jp,
                    names=['id','input/output','type','money','datetime','comment','picture'])

df_ch = df_ch.drop(0)
df_jp = df_jp.drop(0)
```

```python
#抽出当月的数据

#将数据类型转换为日期类型
df_ch['datetime']= pd.to_datetime(df_ch['datetime'])
df_jp['datetime'] = pd.to_datetime(df_jp['datetime'])

#将日期设为索引
df_ch = df_ch.set_index(['datetime'])
df_jp = df_jp.set_index(['datetime'])
```

```python
df_ch2 = df_ch[ym]
df_jp2 = df_jp[ym]

df_ch2
```

```python
#转换日元汇率

#df_jp2['money'] = df_jp2['money']*rate/100

df_jp2['money'] = df_jp2['money'].map(lambda x : float(x)*rate/100)
df_jp2['money'] = df_jp2['money'].round(2)
```

```python
df_ch2['money'] = df_ch2['money'].astype('float64')

df_ch2.info()
```

```python
#合并两张表

df_reslut = df_ch2.append(df_jp2)

df_reslut = df_reslut.drop(columns = ['picture','id'])

df_reslut.describe()
```

```python
#消费行为统计

cash = df_reslut['money'].groupby(df_reslut['input/output'])

cash.sum()
```

```python
#分组统计，排序

cash_flow = df_reslut.groupby(['input/output','type']).sum()

cash_flow.sort_values(['input/output','money'],ascending = False)
```

```python
#导出csv文件
path = 'C:/Users/小奔/Desktop/编程练习/Python/数据分析/learnPython/data/'+ ym +'.csv'

cash_flow.to_csv(path,encoding='gbk')
```

```python
import pandas as pd

path_ch = 'C:/Users/小奔/Desktop/编程练习/Python/数据分析/data/ch_201905.csv'
path_jp = 'C:/Users/小奔/Desktop/编程练习/Python/数据分析/data/jp_201905.csv'

#导入数据
df_ch = pd.read_csv(path_ch,
                    names=['id','input/output','type','money','datetime','comment','picture'])

df_jp = pd.read_csv(path_jp,
                    names=['id','input/output','type','money','datetime','comment','picture'])

df_ch
```

```python
#抽出5月的数据

#将数据类型转换为日期类型
df_ch['datetime']= pd.to_datetime(df_ch['datetime'])
df_jp['datetime'] = pd.to_datetime(df_jp['datetime'])

#将日期设为索引
df_ch = df_ch.set_index(['datetime'])
df_jp = df_jp.set_index(['datetime'])
```

```python
df_ch2 = df_ch['2019-05']
df_jp2 = df_jp['2019-05']

df_ch2
```

```python
#转换日元汇率

rate = 6.3338

df_jp2['money'] = df_jp2['money']*rate/100
df_jp2['money'] = df_jp2['money'].round(2)
```

```python
df_jp2.head(5)
```

```python
#合并两张表

df_201905 = df_ch2.append(df_jp2)

df_201905 = df_201905.drop(columns = ['picture','id'])

df_201905.describe()
```

```python
#消费行为统计

cash = df_201905['money'].groupby(df_201905['input/output']).sum()

cash
```

```python
#分组统计，排序

cash_flow = df_201905.groupby(['input/output','type']).sum()

cash_flow.sort_values(['input/output','money'],ascending = False)
```

```python
#导出csv文件
path = 'C:/Users/小奔/Desktop/编程练习/Python/数据分析/data/201905.csv'

cash_flow.to_csv(path,encoding='gbk')
```

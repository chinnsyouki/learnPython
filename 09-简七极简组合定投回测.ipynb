```python
import xalpha as xa
import pandas as pd

#投资标的
#新华纯债添利A 519152
#易方达沪深300ETF联接A 110020
#嘉实中证500ETF联接  000008
#博时标普500ETF联接A 050025
#大成纳斯达克100指数  000834

aim1 = xa.fundinfo('519152')
aim2 = xa.fundinfo('110020')
aim3 = xa.fundinfo('000008')
aim4 = xa.fundinfo('050025')
aim5 = xa.fundinfo('000834')

#每隔2周定投一次
#data_range = pd.date_range('2018-03-01','2020-03-07',freq='14D')

#每只基金都定投100
#dingtou1 = xa.policy.scheduled(aim1,100,data_range)
#dingtou2 = xa.policy.scheduled(aim2,100,data_range)
#dingtou3 = xa.policy.scheduled(aim3,100,data_range)
#dingtou4 = xa.policy.scheduled(aim4,100,data_range)
#dingtou5 = xa.policy.scheduled(aim5,100,data_range)

#每只基金都进行网格交易
dingtou1 = xa.policy.grid(aim1,[1,1,1,1,1],[5,5,5,5,5],'2019-03-01','2020-03-01',2000)
dingtou2 = xa.policy.grid(aim2,[1,1,1,1,1],[5,5,5,5,5],'2019-03-01','2020-03-01',2000)
dingtou3 = xa.policy.grid(aim3,[1,1,1,1,1],[5,5,5,5,5],'2019-03-01','2020-03-01',2000)
dingtou4 = xa.policy.grid(aim4,[1,1,1,1,1],[5,5,5,5,5],'2019-03-01','2020-03-01',2000)
dingtou5 = xa.policy.grid(aim5,[1,1,1,1,1],[5,5,5,5,5],'2019-03-01','2020-03-01',2000)
```

```python
#调整现金流表
dingtou1.status.set_index('date',inplace=True)
dingtou2.status.set_index('date',inplace=True)
dingtou3.status.set_index('date',inplace=True)
dingtou4.status.set_index('date',inplace=True)
dingtou5.status.set_index('date',inplace=True)

status = pd.concat([dingtou1.status,dingtou2.status,dingtou3.status,dingtou4.status,dingtou5.status],axis=1)

#status.loc['2019-03-01'] = [0,0,0,0,0]
status = status.reset_index()

#status.drop(['050025','000834'],axis = 1,inplace = True)

status
```

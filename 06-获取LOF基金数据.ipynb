```python
import requests
import pandas as pd
import json


#集思录网址LOF基金查询接口）
#股票LOF
#jslUrl = 'https://www.jisilu.cn/data/lof/stock_lof_list/?___jsl=LST___t'
#指数LOF
jslUrl = 'https://www.jisilu.cn/data/lof/index_lof_list/?___jsl=LST___t'

jsl = requests.get(url=jslUrl)

data = jsl.json()

#type(data)

type(data['rows'])
```

```python
list = []

for row in data['rows']:
    #将json数据转换成列表
    list.append(row['cell'])

#创建DataFrame
stocks = pd.DataFrame(data=list)


list = []

for row in data['rows']:
    #将json数据转换成列表
    list.append(row['cell'])

#创建DataFrame
stocks = pd.DataFrame(data=list)


stocks
```

```python
#去除百分比%字符
stocks['discount_rt'] = stocks['discount_rt'].str.replace('%','')
#将字符串转化为数字格式
stocks['discount_rt'] = pd.to_numeric(stocks['discount_rt'],errors='ignore')
stocks['amount_incr'] = pd.to_numeric(stocks['amount_incr'],errors='ignore')
stocks['volume'] = pd.to_numeric(stocks['volume'],errors='ignore')

stocks
```

```python

selected = stocks[(stocks['discount_rt'] >= 2) & 
                  (stocks['apply_status'].str.contains('开放')) & 
                  (stocks['redeem_status'].str.contains('开放')) & 
                  (stocks['amount_incr'] > 0 ) &
                  #(stocks['fund_nm'].str.contains('ETF') == False) &
                  (stocks['volume'] >= 500)]

selected = selected.loc[:,['fund_id','fund_nm','discount_rt','price','estimate_value']]

selected
```
